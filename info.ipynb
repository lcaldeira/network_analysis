```python
import pandas as pd
from datasources import *
```

```python
SyntheticNetworks.setup(base_folder='data', loader_mode='part')
KeggMetabolicNetworks.setup(base_folder='data', loader_mode='part')
SnapEgoNetworks.setup(base_folder='data', loader_mode='part')

data = pd.concat([
    SyntheticNetworks.all(),
    KeggMetabolicNetworks.all(),
    SnapEgoNetworks.all()
], ignore_index=True)
```

```python
TudSocialNetworks.setup(base_folder='data', loader_mode='part')
TudBioinformaticNetworks.setup(base_folder='data', loader_mode='part')

data = pd.concat([
    data, 
    TudSocialNetworks.all(),
    TudBioinformaticNetworks.all()
], ignore_index=True)
```

```python
mask = ((data['N'] == 0) | (data['M'] == 0))
print('empty graphs:', mask.sum())
data[mask].groupby(['collection','dataset']).count()[['source']]
```

```python
data = data[~mask]
data
```

```python
# non-simplified graphs, min-max aggregated values
aggr1 = data.groupby(['collection','dataset'], sort=False).aggregate({
    'label':'nunique',
    'N':['min','max'],
    'M':['min','max'],
    'K':['min','max']
})
aggr1['N'] = aggr1['N'].astype(int)
aggr1['M'] = aggr1['M'].astype(int)

aggr2 = data.groupby(['collection','dataset','label'], sort=False).aggregate({
    'source':'count'
}).groupby(['collection','dataset'], sort=False).aggregate({
    'source':['sum','min','max',]
})

aggr2.join(aggr1).rename({
    'source': 'samples', 
    'label': 'classes', 
    'N':'# nodes', 
    'M':'# edges', 
    'K':'avg. degree'
}, axis=1)
```
